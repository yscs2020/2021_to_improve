```python
from minepy import MINE
m = MINE()
x = np.random.uniform(-1,1,10000)
m.compute_score(x,x**2)
print(m.mic())
def mic(x,y):
    m = MINE()
    m.compute_score(x,y)
    return (m.mic(),0.5)
SelectKBest(lambda X,y:array(map(lambda x:mic(x,Y),x.T)).T,k=2).fit_transform(iris.data,iris.target)
```
